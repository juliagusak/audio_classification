### Audio generation
This is draft notebook to generate audio using pretrained NSynth autoencoder model

```python
import tensorflow as tf
from glob import glob
import numpy as np
import os

import nsynth_utils
```

##### Set postfix = :
    ".npy" to generate audio from encoding (latent vector)
    
    ".wav" if latent vector is unknown. It will first obtain latent representation from initial signal and then reconstruct signal from it. 

```python
postfix = '.npy'
```

##### Specify path to the
    source folder (with '.npy' files if you want to generate audio from encodings (one file contains encoding for one audio)  or  to the folder with '.wav' files)
    
    folder with generated .wav files
    
    pre-trained model

```python
source_path = '/home/julia/DeepVoice_project/saved_data/samples/'
save_path = '/home/julia/DeepVoice_project/saved_data/samples/'
checkpoint_path = "/home/julia/magenta-demos/jupyter-notebooks/wavenet-ckpt/model.ckpt-200000"

sample_length = 40000
batch_size = 1
```

##### Find all files with specified extention in the source folder

```python
if postfix == ".wav":
    files = sorted(glob('{}/**/*.wav'.format(source_path), recursive=True))
elif postfix == ".npy":
    files = sorted(glob('{}/**/*.npy'.format(source_path), recursive=True))

print('Source folder contains {} files'.format(len(files))) 
print('...{}'.format(files[-3:]))
```

##### Run generation process and save the result
to use GPU set gpu_number = 1 and specify gpu_name

```python
# Now synthesize from files one batch at a time
# samples_per_save = 1000
gpu_number = 0
gpu_name = "/device:GPU:{}".format(gpu_number)

for start_file in range(0, len(files), batch_size):
    end_file = start_file + batch_size
    files_batch = files[start_file:end_file]
    save_names = [os.path.join(save_path, 
                               "gen_" + os.path.splitext(os.path.basename(f))[0]) + ".wav"
                  for f in files_batch]
    batch_data = nsynth_utils.load_batch(files_batch, sample_length = sample_length)
    
    print(batch_data.shape)
    encodings = batch_data if postfix == ".npy" else nsynth_utils.encode(batch_data, checkpoint_path, sample_length=sample_length)
    if gpu_number != 0:
        with tf.device("{}".format(gpu_name)):
            nsynth_utils.synthesize(encodings, save_names, checkpoint_path) 
    else:
        nsynth_utils.synthesize(encodings, save_names, checkpoint_path)
```
